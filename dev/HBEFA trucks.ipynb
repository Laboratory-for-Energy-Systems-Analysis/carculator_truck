```python
fp = r"C:\Users\romai\Documents\GitHub\carculator_truck\dev\EFA_HOT_Subsegm_HGV.XLSX"
df = pd.read_excel(fp)
df.head()
```

```python
df["Subsegment"].unique().tolist()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
rcParams.update({'figure.autolayout': True})

fig, axs = plt.subplots(figsize=(12,10), 
                        nrows=4, ncols=3,     # fix as above
                        gridspec_kw=dict(hspace=0.2),
                        sharey=True) # Much control of gridspec

powertrains = [
    #"diesel",  
    "CNG"
]


subsegments = [
    #"HGV CNG <=7,5t Euro-IV",
    #"HGV CNG <=7,5t Euro-V",
    #"HGV CNG <=7,5t Euro-VI",
    #"HGV CNG >7,5-12t Euro-IV",
    #"HGV CNG >7,5-12t Euro-V",
    #"HGV CNG >7,5-12t Euro-VI",
    #"HGV CNG >12t Euro-IV",
    #"HGV CNG >12t Euro-V",
    #"HGV CNG >12t Euro-VI",
    #"TT/AT CNG Euro-IV",
    #"TT/AT CNG Euro-V",
    #"TT/AT CNG Euro-VI"
    "TT/AT >20-28t Euro-VI"
]

components = df['Component'].unique().tolist()


euro_classes = df["EmConcept"].unique().tolist()

sizes = [
'RT ≤7,5t',
'RT >7,5-12t',
'RT >12t',
'RT >14-20t',
'RT >26-28t',
'TT/AT >40-50t',
'TT/AT >50-60t',
'TT/AT not specified'
]


idx=pd.IndexSlice

l_res = []

for sub in subsegments:
    for component in components:
        data = df.loc[(df["Subsegment"]==sub)
                      &(df["Component"]==component), ["V", "EFA"]]

        ind = data["V"].values
        vals = data["EFA"].values

        if len(data)>0:

            z = np.polyfit(ind, vals, 3)
            f = np.poly1d(z)

            l_res.append([sub, component, f[0], f[1], f[2], f[3]])
            #print(component, f, z)

            # calculate new x's and y's
            x_new = np.linspace(0, 150, 50)
            y_new = np.clip(f(x_new), 0, None)

        ax = plt.subplot(5, 3, components.index(component)+1)
        ax = plt.subplot(5, 3, components.index(component)+1)

        ax.plot(ind, vals, marker=".", linestyle="")
        ax.plot(x_new, y_new,  linestyle="-",  label=sub)
        ax.set_title(component)

        if components.index(component)==12:
            ax.legend(ncol=3)

#pd.DataFrame(l_res, columns=["powertrain", "euro_class", "size", "component", "intercept", "c", "b", "a"]).to_excel("HEM_factors_trucks.xlsx")
```
